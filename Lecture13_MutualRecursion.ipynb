```python
# Mutual Recursion
def isEven(num):
    if num == 0: #base case
        return True

    return isOdd(num - 1)


def isOdd(num):
    if num == 0: # base case
        return False

    return isEven(num - 1)

print(isOdd(1))
print(isEven(2))
print(isEven(7))
```

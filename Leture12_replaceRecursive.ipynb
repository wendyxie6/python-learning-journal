```python
def writeVertical(num):
    if num < 0:
        print("-")
        num *= -1

    numDigits = 1
    temp = num

    while temp >= 10:
        temp //= 10
        numDigits += 10

    while numDigits > 0:
        digit = num // numDigits
        print(digit)

        num %= numDigits
        numDigits //= 10

writeVertical(1234)
```

# **Combining DataFrames in Pandas (Joins & Concatenation)**

```python
import pandas as pd

customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 104, 105],
    'name': ['Alice', 'Bob', 'Carlos', 'Diana', 'Evan'],
    'city': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Seattle'],
    'loyalty_tier': ['Gold', 'Silver', 'Gold', 'Bronze', 'Silver']
})

orders = pd.DataFrame({
    'order_id': [1001, 1002, 1003, 1004, 1005, 1006],
    'customer_id': [101, 103, 101, 104, 106, 103],
    'order_total': [250.5, 99.9, 320.0, 45.0, 180.0, 75.5],
    'order_date': pd.to_datetime([
        '2024-01-05', '2024-01-10', '2024-02-02',
        '2024-02-14', '2024-02-20', '2024-03-01'
    ])
})

shipments = pd.DataFrame({
    'cust_id': [101, 103, 104, 105],
    'carrier': ['UPS', 'FedEx', 'USPS', 'UPS'],
    'shipping_cost': [15.0, 12.5, 8.0, 10.0]
})
```

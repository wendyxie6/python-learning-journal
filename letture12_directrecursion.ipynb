```python
# Recursive search - direct recursion
# Base case: we find out answer
# Base case 2:  we run out of space to search
# Recursive case: split list our list i half, if answer not present at middle, search in the correct direction recursively
def recursiveSearch(array, target, bottom, top):
    mid = (top + bottom) // 2
    if array[mid] == target:
        return mid
    elif target < array[mid]:
        return recursiveSearch(array, target, bottom, mid - 1)
    else:
        return recursiveSearch(array, target, mid + 1, top)

mylist = [1, 2, 3, 4, 5, 6, 5]

recursiveSearch(mylist, 6, 0, len(mylist) - 1)
```

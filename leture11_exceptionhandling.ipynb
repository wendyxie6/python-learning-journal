```python
def isFloatable(value):
    try:
        float(value)
        #return True - valid return
    except ValueError as e:
        print(f"Error, could not conver  {value} into a float")
        return False
    else:
        return True

    # Retun True - Valid return

print(isFloatable("Five"))
print(isFloatable("12"))

percent = input("enter percentage as a fraction between 0 and 1: ")

while(not isFloatable(percent) or float(float(percent)> 1 or float(percent) < 0)):
    print(f"Error: {percent} is not a fraction between 0 and 1")

    percent = input("Please enter a valid value: ")

```

## â˜„ï¸ **1. Projectile Motion â€” Experimental vs Theoretical**

A projectileâ€™s vertical position is given by:

**$ y = v_0 t - \frac{1}{2} g t^2 $**

where:  
- $ v_0 = 20\,m/s, g = 9.8\,m/s^2$

---

### **ðŸ§© Task**
1. Generate `t` values from 0 to 4 seconds (0.2 s step).  
2. Compute the **theoretical** height `y_theoretical` using the above formula.  
3. Simulate **experimental** measurements by adding small random noise to it using `np.random.normal()`.  
4. Plot:
   - Experimental data as a **scatter plot** (red markers).  
   - Theoretical curve as a **smooth line** (blue line).

### ðŸ’¡ *Hint*

**Note:** NumPyâ€™s trigonometric functions (np.sin, np.cos, np.tan) expect angles in radians, not degrees.

Screen Shot 2025-10-17 at 1.54.04 PM.png

````python
import matplotlib.pyplot as plt
import numpy as np

u = 20
g = 9.8

deg = np.radians(45)

````

````python
t = np.arange(0., 4.2, 0.2)
x = u * np.cos(deg) * t
y = u*t - g * (t**2)/2

plt.plot(x, y, label="Theoretical Trajectory")

noise = np.random.normal(0, 1, len(t))
y_exp = y + noise


plt.scatter(x, y_exp, color = "red", label="Experimental Data (with noise)")

plt.title("Project Motion: Theoretical va Experimental")
plt.xlabel("Horizontal Distance (m)")
plt.ylabel("Vertical Height(m)")
plt.grid(True, linestyle='-', alpha=1)
plt.legend(loc="lower left")
````

## **Social Media Engagement Over Time**

In this exercise, you will practice using **Matplotlib subplots** to visualize two related aspects of social media data:

1. **Monthly average engagement** (likes per post) â€” using a **bar plot**.  
2. **Daily engagement trend** (for one month) â€” using a **line or scatter plot**.

This exercise helps you connect:
- **Aggregated data** (monthly summaries)
- **Continuous data** (daily time trends)
- **Layout skills** with `plt.subplots()`

---

### **Tasks**

1. You are provided with the following data:
   - month_names: Names of the 12 months
   - monthly_engagement: Average likes per post for each month
   - days_june: Days of June (1â€“30)
   - engagement_june: Likes per post for each day in June

2. Using this data, create a subplot figure with 1 row and 2 columns:
   - Left subplot: Bar plot of average engagement per month
   - Right subplot: Scatter or line plot of daily engagement trend for June
3. Customize your plots:
   - Add titles and axis labels
   - Include gridlines for better readability
   - Add a legend to the daily trend plot
4. Make the overall figure visually appealing and easy to interpret.


````python
import numpy as np
import matplotlib.pyplot as plt

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18,6))

# --- 1) Generate synthetic data ---
np.random.seed(42)
months = np.arange(1, 13)
month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']

# Average monthly engagement (likes per post)
monthly_engagement = np.random.randint(200, 800, size=12)

axes[0].bar(months,monthly_engagement)
axes[0].set_title("Average Engagement Per Month")
axes[0].set_xlabel("Month")
axes[0].set_xticks(months, month_names)
axes[0].set_ylabel("Average Likes Per Post")
axes[0].grid(axis="y", linestyle="--", linewidth=0.5)

# Daily engagement data for June (30 days)
days_june = np.arange(1, 31)
# Simulate weekly pattern (weekend peaks) + noise
engagement_june = 400 + 150*np.sin(days_june/4) + np.random.normal(0, 40, size=30)


axes[1].plot(days_june,engagement_june, color = "red", marker='o',ms=10)
axes[1].set_title("Daily Engagement Trend (June)")
axes[1].set_xlabel("Day of Month")
axes[1].set_ylabel("Likes Per Post")
axes[1].grid(True, linestyle="--", linewidth=0.5)




````

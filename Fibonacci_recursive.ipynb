```python
def fibonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

# Test the function
num = int(input("Enter how many Fibonacci numbers to display: "))

print("Fibonacci sequence:")
for i in range(num):
    print(fibonacci(i), end=" ")
```

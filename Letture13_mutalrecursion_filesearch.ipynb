```python
import os

def processFolder(path):
    print(f"[Directory] {path}")

    for item in os.listdir(path):
        fullPath = path + "/" + item

        processItem(fullPath)

def processItem(path):
    if os.path.isfile(path):
        print(f"[File] {path}")
    else:
        processFolder(path)

path = "/Users/wendy/python_wv"
processFolder(path)
```

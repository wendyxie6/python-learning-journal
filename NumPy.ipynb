## **Shape**

Shape is an attribute that arrays have (not a method)

```python
arr_1d = np.arange(15)
arr_1d
```

```python
# Vector
arr_1d.shape
```

means:

  * one dimension
  * length = 15

The comma matters.

In NumPy, a **1D array has no concept of row vs column**, but you can reshape it into a 2D column vector.

***No, (,15) does NOT exist in NumPy.***

**If you want a row vector:**

```python
# Notice the two sets of brackets
arr_2d = arr_1d.reshape(1,15)
arr_2d
```

```python
arr_2d.shape
```

**If you want a column vector:**

```python
arr_2d_col = arr_2d.reshape(15,1)
arr_2d_col
```

```python
arr_2d_col.shape
```

```python
arr_2d.reshape(-1,1)
```

| Shape     | Meaning                    |
| --------- | -------------------------- |
| `(15,)`   | 1D array with 15 elements  |
| `(15, 1)` | 2D column vector           |
| `(1, 15)` | 2D row vector              |
| `(,15)`   | ❌ invalid / does not exist |



* A shape must specify all dimensions explicitly.

* A 1D array has only one axis, so it cannot be a row or a column.

* NumPy does not distinguish row vector vs column vector for 1D arrays.

`reshape()` **does NOT modify the array in place** — it returns a new array (or a new view).

**Views vs copies**

Does it return something?
    
  * YES → not in place → must assign
  * NO  → in place

## **Broadcasting**

NumPy arrays differ from normal Python lists because of their ability to broadcast. With lists, you can only reassign parts of a list with new parts of the same size and shape. That is, if you wanted to replace the first 5 elements in a list with a new value, you would have to pass in a new 5 element list. With NumPy arrays, you can broadcast a single value across a larger set of values:

```python
#Setting a value with index range (Broadcasting)
myArray[0:5]=100
myArray
```

```python
myArray = np.arange(15)
myArray
```

```python
#Important notes on Slices
sliceArray = myArray[0:4]
sliceArray
```

```python
#Change Slice
sliceArray[:]=99
sliceArray
```

Now note the changes also occur in our original array!




```python
myArray
```

Because NumPy slicing returns a **view**, not a copy — and views share the same underlying memory.

Data is not copied, it's a view of the original array!

```python
#To get a copy, need to be explicit
myArray_copy = myArray.copy()
myArray
```

This is intentional (performance reason)

NumPy is designed for:

  * large arrays
  * scientific computing

Copying data by default would be:

  * slow
  * memory-expensive

So slicing gives a view by default.

## **Arithmetic**


```python
import numpy as np
arr = np.arange(0,10)
arr
```

```python
arr + arr
```

```python
arr - arr
```

```python
arr * arr
```

NumPy arrays are numerical objects

 * `np.array` represents a vector of numbers
 * Arithmetic operators are **element-wise** by default

**Design philosophy:** Operators mean element-wise math unless stated otherwise

**Numpy vs Matlab:** NumPy assumes numbers first; MATLAB assumes matrices first.

```python
arr**3
```

**Numpy vs Python in handling division by zero:**



```python
# This will raise a Warning on division by zero, but not an error!
# It just fills the spot with nan
arr/arr
```

```python
# Also a warning (but not an error) relating to infinity
1/arr
```

```python
# Pure Python raises an error
1 / 0
```

```python
# Pure Python raises an error
0/0
```

Why?

  * Python treats numbers as exact values
  * Division by zero is considered a logic error
  * Execution stops immediately

This is a language-level safety rule.

NumPy follows the IEEE 754 floating-point standard, which defines:

| Operation       | Result          |
| --------------- | --------------- |
| `x / 0` (x ≠ 0) | `inf` or `-inf` |
| `0 / 0`         | `nan`           |
| `inf - inf`     | `nan`           |

This is critical for:

  * vectorized computation
  * simulations
  * optimization

avoiding crashes in long-running numerical code

Why this difference matters (key insight)

  * **Python** → correctness-first, fail fast
  * **NumPy** → performance & numerical analysis, fail gracefully

Stopping a 10-million-element computation because one value was zero would be disastrous.

# **NumPy vs Python List**




```python
import numpy as np
import time

N = 10_000_000
pList = list(range(N))
nArray = np.arange(N)
```

### **Pure Python**

```python
start = time.time()
result = []
for i in pList:
    result.append(i * 2)
print("For loop:", time.time() - start)
```

### **List comprehension**

```python
start = time.time()
result = [i * 2 for i in pList]
print("List comprehension:", time.time() - start)
```

### **NumPy**

```python
start = time.time()
result = nArray * 2
print("NumPy:", time.time() - start)
```

####**Key Observations:**

  🐢 **For loop:** Works but slow & verbose

  🐍 **List comprehension:** Cleaner, but still Python-level speed

  ⚡ **NumPy:** Allows vectorized operations (no loops needed), Runs in optimized C under the hood → Shortest, fastest, suitable for ML

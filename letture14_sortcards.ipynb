```python
def selectionSort(arr):
    #iterate from the start of the list until just before the last item
    for x in range(len(arr) - 1):
        #look for the smallest element in unsorted portion of the list(the list from x till end)
        smallest = x

        for y in range(x, len(arr)):
            if arr[smallest] > arr[y]:
                smallest = y
        # once we have the smallest element, swap it for the element at x

        swap = arr[smallest]
        arr[smallest] = arr[x]
        arr[x] = swap

    #print(arr)

list_arr = [3, 15, 8, 10]

selectionSort(list_arr)

print(list_arr)

```

```python
class Car:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def print(self):
        print(f"you are driving a {self.make} {self.model}")


class Motorcycle:
    def __init__(self, make, model):
        self.make = make
        self.model = model

    def print(self):
        print(f"Motorcyle stats: {self.make} {self.model}")

c1 = Car("Honda", "Civic")
m1 = Motorcycle("Suzuki", "Razer")

array = [c1, m1]
#print(array)

for item in array:
    item.print()
```

```python
class Card:
    def __init__(self, color, ability):
        self.color = color
        self.ability = ability

    def use(self, topCard):
        pass

class NumCard(Card):
    def __init__(self, color, ability, number):
        super()._init__(color, "None")
        self.number = number

    def use(self, topCard):
        if topCard.ability == "None":
            if self.color == topCard.color or self.number == topCard.number:
                return True
            else:
                return False

        if topCard.ability == "Wild":
            if self.color == topCard.color:
                return True
            else:
                return False

class WildCard(Card):
    def __init__(self):
       super().__init__("Wild", "Wild")

    def use(self, topCard):
        if topCard is not None:
            self.color = input("Please select a color for your wild card!")
            return True


hand = [NumCard("blue", 3), NumCard("yellow", 0), WildCard()]
topCard = NumCard("yellow", 7)

for card in hand:
    if card.use(topCard):
        print(f"You can play this card {card}")


```
